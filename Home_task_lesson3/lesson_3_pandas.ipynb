# *What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)*

````python
Top15.isna().value_counts()
````

````python
def answer_two():
    Top15 = answer_one()
    avgGDP1 = Top15.loc[:, '2006':'2015'].mean(axis=1)  #Середнє 2006 - 2015
    avgGDP1 = avgGDP1.sort_values(ascending=False)       #По спаданню 

    return avgGDP1

answer_two()
````

This function should return a tuple with the name of the country and the ratio.

````python
def answer_four():
    Top15 = answer_one()
    Top15['Ratio_Cit'] = Top15['Self-citations'] / Top15['Citations']

    max_value = Top15['Ratio_Cit'].max()
    conrty_with_highest_r = Top15['Ratio_Cit'].idxmax()
    return max_value, conrty_with_highest_r
````

# Question 5
## *Create a column that estimates the population using Energy Supply and Energy Supply per capita. What is the third most populous country according to this estimate?*

This function should return a single string value.

````python
def answer_five():
    Top15 = answer_one()
    Top15['the_population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    sort_by_population = Top15['the_population'].sort_values(ascending=False)
    the_third_most_populous = sort_by_population.index[2]
    return the_third_most_populous

print(answer_five())
type(answer_five())
````

# Question 6
 ## *Create a column that estimates the number of citable documents per person. What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the .corr() method, (Pearson's correlation).*

This function should return a single number.

````python
def answer_six():
    Top15 = answer_one()
    Top15['number_of_doc'] = Top15['Citable documents'] / Top15['the_population']
    corr = Top15['number_of_doc'].corr(Top15['Energy Supply per Capita'])
    return corr

answer_six()
````

# Question 7
*Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.*

This function should return a DataFrame with index named Continent ['Asia', 'Australia', 'Europe', 'North America', 'South America'] and columns ['size', 'sum', 'mean', 'std']

````python
def answer_seven():
    Top15 = answer_one()

    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    
    
    Top15['Continent'] = Top15.index.map(ContinentDict)        # метод .map уявно перекладає відвовідні країни у відповідні континенти і використовує як індекс
    for_df_stats = Top15.groupby('Continent')['the_population'].agg(['size', 'sum', 'mean', 'std']) #Групуємо та методом .agg обчислюємо статистики
    ordered_continents = ['Asia', 'Australia', 'Europe', 'North America', 'South America'] #так легше
    for_df_stats = for_df_stats.reindex(ordered_continents)
    return for_df_stats

# в колонці std є NaN, бо в деяких континентах не вистарчає значень
````
