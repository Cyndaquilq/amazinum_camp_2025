```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
df = pd.read_csv('./data/Car_Insurance_Claim.csv', index_col=0)
df.head(15)
```

```python
print(df['ID'].duplicated().value_counts())           # Логічно пребположити, що айді унікальне має бути

print(df['POSTAL_CODE'].value_counts())                 #Тут в деяких флоат значення
```

```python
# print(df['SPEEDING_VIOLATIONS'].value_counts())
# print(df['DUIS'].value_counts())
# print(df['PAST_ACCIDENTS'].value_counts())

print(df['OUTCOME'].value_counts())
```

```python
df = df.fillna(value=np.nan)   # none заміняєм на NaN
df.describe(include='all')
df.isna().sum()
```

```python
df
```

```python
df.describe(include='all')
df.isna().sum()
```

```python
print(df.dtypes)

df.head(5)
```

```python
columns_on_num = ['VEHICLE_OWNERSHIP','MARRIED', 'CREDIT_SCORE', 'CHILDREN','ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 
                         'DUIS', 'PAST_ACCIDENTS',   
                          'POSTAL_CODE', 'OUTCOME']


for column in columns_on_num:
    df[column] = pd.to_numeric(df[column], errors='coerce')

print(df.dtypes)
```

```python
for col in df.select_dtypes(include=['float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# То лише варнінги
```

```python
df.isna().sum()  # файно
```

```python
evil_df = df[['SPEEDING_VIOLATIONS',	'DUIS',	'PAST_ACCIDENTS','OUTCOME']]
evil_df['SUM_ACC'] = evil_df['SPEEDING_VIOLATIONS'] + evil_df['DUIS'] + evil_df['PAST_ACCIDENTS']
evil_df
```

```python
evil_df['SPEEDING_VIOLATIONS'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(evil_df['SUM_ACC'], bins=10, kde=True, color='royalblue')
plt.title("Розподіл кількості порушень SUM_ACC")
plt.xlabel("Кількість порушень")
plt.ylabel("Кількість людей")
plt.show()
```
